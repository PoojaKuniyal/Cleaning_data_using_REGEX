## Data cleaning

```python
df['Product_Sold | Description'].head()
```

```python
df['Product_Sold | Description'] = df['Product_Sold | Description'].str.replace("â€", "")

df['Product_Sold | Description'] = df['Product_Sold | Description'].str.replace("آ®آ","")

df['Product_Sold | Description'] = df['Product_Sold | Description'].str.replace("®","")

df['Product_Sold | Description'] = df['Product_Sold | Description'].str.replace("™", "")

df['Product_Sold | Description'] = df['Product_Sold | Description'].str.replace(",",'-')

```

```python
df['Product_Sold | Description'].head()
```

```python
# # Make Brand and use_case column by extracting it from Product_Sold | Description column


#df['Product Name'] = df['Product'].str.extract(r'([A-Za-z0-9\s\(\)\+\-]+)(?= (?:PHANTEKS|915 Generation|USB|DX01|A09P))', expand=False)

# Define the regular expressions for  brand, and use case
brand_pattern = r'(DELL|Lenovo|Asus|Apple|HP|TAGITOP|Generic|Gigabyte|Microsoft|Fujitsu|ITX|MSI|Intel Core i7|Intel Core i9)'

use_case_pattern = r'(gaming|laptop|HTPC|desktop|mini ITX|Raspberry Pi|power adapter|stand|headset|cooling case|Phone holder|Speaker)'

# Extracting Brand and Use Case from the Product_Sold | Description column
df['Brand'] = df['Product_Sold | Description'].str.extract(brand_pattern, expand=False, flags=re.IGNORECASE)
df['Use_Case'] = df['Product_Sold | Description'].str.extract(use_case_pattern, expand=False, flags=re.IGNORECASE)

# Fill missing values in the Brand and use_case column with "Unknown"
df['Brand']= df['Brand'].fillna("Unknown")
df['Use_Case'] = df['Use_Case'].fillna("Unknown")
pd.reset_option('display.max_colwidth')
df.head()
```

```python
# Make RAM and ROM column (Exract it from Product_Sold | Description column)
# Define a regular expression pattern to extract RAM and ROM information

ram_pattern = r"(\d+GB |\d+ GB |\d+ GB)"

rom_pattern = r"(\d+\s*(?:TB|GB)?\s*(?:HDD|SSD))"

# Extract the RAM information and create a new RAM column
df['RAM'] = df['Product_Sold | Description'].str.extract(ram_pattern, flags= re.IGNORECASE)
df["RAM"] = df["RAM"].str.replace("GB", " GB")
df['RAM'] = df['RAM'].fillna('unknown')

df['ROM'] = df['Product_Sold | Description'].str.extract(rom_pattern, flags= re.IGNORECASE)
df['ROM'] = df['ROM'].fillna('unknown')

df.head()
```

```python
# cleaning old and current price column and convert them to numeric

df['Old Price'] = df['Old Price'].str.replace(r'[^\d.]', '', regex=True)

df['Current Prices'] = df['Current Prices'].str.replace(r'[^\d.]','', regex=True)

# Replace empty values with 0
df["Old Price"] = df["Old Price"].replace('', '0')
df["Current Prices"] = df["Current Prices"].replace('', '0')

# Convert the columns to float
df["Old Price"] = df["Old Price"].astype(float)
df["Current Prices"] = df["Current Prices"].astype(float)
df.head()
```

```python
df["Discount Percentages"].unique()
```

```python
# Convert Discount Percentages column to percentages

df["Discount Percentages"] = (df["Discount Percentages"] * 100) # its a float column

df.head()
```

```python
df['Rating'].unique()
```

```python
# Rename the "Rating" column to "Rating_out_of_5"
df.rename(columns={"Rating": "Rating_out_of_5"}, inplace=True)

# Convert the column to string type, handle missing values, and then clean it
df["Rating_out_of_5"] = df["Rating_out_of_5"].astype(str)  # Ensure the column is string
df["Rating_out_of_5"] = df["Rating_out_of_5"].replace("No ratings available", "0")  # Handle missing ratings
df["Rating_out_of_5"] = df["Rating_out_of_5"].str.replace(" out of 5", "")  # Remove " out of 5"
df["Rating_out_of_5"] = df["Rating_out_of_5"].astype(float)  # Convert to float after cleaning

df.head(3)
```

```python
df["Number of Ratings"].unique()
```

```python
# Cleaning the Number of Ratings column

# Remove " (" and ")" from the Number of Ratings column and replace other text
df["Number of Ratings"] = df["Number of Ratings"].str.replace(")", "").str.replace("(", "").str.replace("No ratings available", "0").str.replace(" verified rating", "")


# Convert the Number of Ratings column to integers and handle unexpected values
df["Number of Ratings"] = pd.to_numeric(df["Number of Ratings"], errors='coerce').fillna(0)

df.head()
```

```python
# Cleaning the Unit Available column

# Remove " units left" from the "Unit Available" column and replace other text
df["Unit Available"] = df["Unit Available"].str.replace(" units left", "")

# Convert the "Unit Available" column to integers and handle unexpected values
df["Unit Available"] = pd.to_numeric(df["Unit Available"], errors='coerce').fillna(0).astype(int)

df.head()
```

```python
# Create a new column with prices in USD
df['USD Price'] = df['Current Prices'] / 30.90

# Round the USD prices to two decimal places
df['USD Price'] = df['USD Price'].round(2)

df.head()
```

```python
# see the describtion of the Numeric columns in the dataframe

print("Summary Of The Dataset with Numeric columns :")
df.describe()
```

```python
# cleaning Offer Type column

# Rename the Offer Type column to Sale Type
df = df.rename(columns={'Offer Type': 'Sale Type'})
print(df['Sale Type'].unique())


# Remove "Offer Type:" and whitespaces from the Sale Type column
df['Sale Type'] = df['Sale Type'].str.replace("Offer Type: ", "").str.strip()
df['Sale Type'] = df['Sale Type'].str.replace("Offer type:", "").str.strip()

print(df['Sale Type'].unique())
```

```python
df['Seller Name and Score'].sample(3)
```

```python
# Define a custom function to split and extract the seller name
def extract_seller_name(s):
    if pd.notna(s) and 'Seller Name:' in s:
        return s.split('Seller Name: ')[1].split(',')[0].strip()
    else:
        return None

# Define a custom function to extract the Seller score and handle missing values

def extract_score(s):
    if pd.notna(s) and 'Score:' in s:
        score_text = s.split('Score: ')[1]
        if '%' in score_text:
            return float(score_text.strip('%'))
    return None


df['Seller Name'] = df['Seller Name and Score'].apply(extract_seller_name)
df['Seller Score'] = df['Seller Name and Score'].apply(extract_score)
df = df.drop(columns=['Seller Name and Score'])

df[['Seller Name', 'Seller Score' ]]
```

```python
# Split the "Door Delivery Info" column into three columns
df[['Door Delivery Fees', 'Door Delivery Date', 'Door Delivery Ready Time']] = df['Door Delivery Info'].str.split(', ', n=2, expand=True)

# Remove unwnanted words from each column
df['Door Delivery Fees'] = df['Door Delivery Fees'].str.replace("Delivery Fees: ", "")
df['Door Delivery Date'] = df['Door Delivery Date'].str.replace("Delivery Dates: ", "")

df['Door Delivery Ready Time'] = df['Door Delivery Ready Time'].str.replace("Ready Time: ", "")

df = df.drop(columns=['Door Delivery Info'])

df.head(3)
```

```python
# Split the "Pickup Station Info" column into three columns
df[['Pickup Fees', 'Pickup Date', 'Pickup Ready Time']] = df['Pickup Station Info'].str.split(', ', n=2, expand=True)

# Remove prefixes from each column
df['Pickup Fees'] = df['Pickup Fees'].str.replace("Pickup Fees: ", "")
df['Pickup Date'] = df['Pickup Date'].str.replace("Pickup Dates: ", "")
df['Pickup Ready Time'] = df['Pickup Ready Time'].str.replace("Ready Time: ", "")

df = df.drop(columns=['Pickup Station Info'])

df.head(3)
```

```python
# Drop the "Return Policy Info" column
df = df.drop(columns=['Return Policy Info'])
```
